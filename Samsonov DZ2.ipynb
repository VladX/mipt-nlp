```python
import pandas

data = pandas.read_csv('hillary-clinton-emails/Emails.csv')
```

```python
data.columns
```

```python
data['RawText'].head()
```

```python
import re

data['RawText'][0].replace('\n',' ')

def preprocess(t):
    t = t.replace('\n', ' ')
    t = t.replace('@', ' ')
    t = t.replace('&', ' ')
    t = t.replace('-', ' ')
    t = t.replace('+', ' ')
    t = t.replace('_', ' ')
    t = t.replace('UNCLASSIFIED', ' ')
    t = t.translate(None, ',!.;()?\'":<>/$*=%')
    t = re.sub(r'[0-9]+', ' ', t)
    t = re.sub(r' +', ' ', t)
    return t.lower()

texts = data['RawText']
texts = map(preprocess, texts)
texts[0]
```

В тексте все специальные символы заменены на пробелы.

Все знаки препинания удалены, а сам текст приведен к нижнему регистру. Это сделано для того, чтобы при делении текста на слова и биграммы не возникало мусорных сочетаний вида СЛОВО+ЗНАК, БУКВА+ЗНАК (для биграмм). Нижний регистр нужен, чтобы слова разными регистрами считались за одно слово.

```python
from sklearn.feature_extraction.text import CountVectorizer
from collections import defaultdict
import operator

an = CountVectorizer(ngram_range=(2, 2)).build_analyzer()
count = defaultdict(int)
for t in texts:
    for bi in an(t):
        count[bi] += 1

count = sorted(count.items(), key=operator.itemgetter(1), reverse=True)
print count[:10]
```

```python
import nltk
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
words = []
for t in texts:
    words += t.split(' ')
finder = BigramCollocationFinder.from_words(words)
finder.nbest(bigram_measures.pmi, 10)
```

Будем визуализировать крастеризацию с помощью понижения пространства до двумерного через **TruncatedSVD**.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering
import numpy as np

vec = TfidfVectorizer(min_df=20, max_df=8000).fit_transform(texts)
print(vec.shape)
```

### Кластеризация через KMeans

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD

km = KMeans(3).fit(vec)
vis = TruncatedSVD().fit_transform(vec)

plt.figure(figsize=(10,10), dpi=500)
colors = plt.cm.hot(np.linspace(0, 1, km.cluster_centers_.shape[0]))
for i in xrange(km.cluster_centers_.shape[0]):
    idx = np.where(km.labels_ == i)
    plt.scatter(vis[idx][:,0], vis[idx][:,1], c=colors[i])
plt.show()
```

### Кластеризация через AgglomerativeClustering

К сожалению, **AgglomerativeClustering** не поддерживает sparse-матрицы, полученные после TF-IDF (сделать из нее dense-матрицу нельзя, т.к. памяти нужно очень много).

Поэтому делаем так:

1. Сначала понижаем размерность до 400.
2. Потом кластеризуем с dense-матрицой.
3. Снова понижаем размерность до 2 для визуализации.

```python
low_dim_vec = TruncatedSVD(400).fit_transform(vec)
km = AgglomerativeClustering(3).fit(low_dim_vec)
vis = TruncatedSVD().fit_transform(low_dim_vec)

plt.figure(figsize=(10,10), dpi=500)
colors = plt.cm.hot(np.linspace(0, 1, 3))
for i in xrange(3):
    idx = np.where(km.labels_ == i)
    plt.scatter(vis[idx][:,0], vis[idx][:,1], c=colors[i])
plt.show()
```

### Кластеризация через SpectralClustering

```python
low_dim_vec = TruncatedSVD(400).fit_transform(vec)
km = SpectralClustering(3).fit(low_dim_vec)
vis = TruncatedSVD().fit_transform(low_dim_vec)

plt.figure(figsize=(10,10), dpi=500)
colors = plt.cm.hot(np.linspace(0, 1, 3))
for i in xrange(3):
    idx = np.where(km.labels_ == i)
    plt.scatter(vis[idx][:,0], vis[idx][:,1], c=colors[i])
plt.show()
```

## Оценка интерпретируемости кластеров с помощью ассессоров

Можно выбрать по $m$ случайных примеров из каждого из $k$ кластеров. Если у кластера есть центроид, можно дополнительно потребовать, чтобы примеры были не дальше некоторого расстояния от центра своего кластера (чтобы не показывать ассессорам пограничные примеры).

Полученные $mk$ примеров показываем $n$ ассессорам. Предлагаем им разделить $mk$ примеров на $k$ кластеров. Обозначим за $a_i$ разметку ассессора, за $b_i$ -- разметку алгоритма. Введем метрику качества для одного ассессора:

$$q(a,b)=\max_{\sigma(k)} \frac{1}{mk} \sum_{i=1}^{mk} I(a_{\sigma(i)} = b_i)$$

берется максимум по всем возможным перестановкам из $k$ элементов.

Итоговое качество для $n$ ассессоров — это среднее качество по всем ассессорам:

$$Q(a,b) = \frac{1}{n} \sum_{j=1}^{n} q(a^j,b^j)$$

### Оценка качества кластеризации KMeans

Был проведен эксперимент с $n=3$ ассессорами и $m=2$ примерами из каждого кластера.

Качество по введенной метрике получилось таким:

$Q = \frac{1}{3}(1 + \frac{2}{3} + \frac{2}{3}) = 0.7777777777777777... \approx 0.78$
